```python
import matplotlib.pyplot as plt
import numpy as np
def NormalizedSpecificWork(temperature_ratio,pressure_ratio):
  gamma = 1.4
  a = 1 - (pressure_ratio**(-1*(gamma-1)/gamma))
  b = pressure_ratio**((gamma-1)/gamma) - 1
  return temperature_ratio*a - b
pressure_ratio_values = np.linspace(1,40,100)
number_of_temperature_ratios = 5
starting_temperature_ratio = 1
temperature_ratio_values = np.arange(starting_temperature_ratio,starting_temperature_ratio+2*number_of_temperature_ratios,2)
for i in temperature_ratio_values:
  plt.plot(pressure_ratio_values,NormalizedSpecificWork(i,pressure_ratio_values),label=f't = {i}')
plt.legend()
plt.xlabel('Pressure Ratio')
plt.ylabel('Work')
plt.title('Normalized Specific Work')
plt.show()
```
