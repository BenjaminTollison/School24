```python
bgr = plt.imread("BGR.jpg")
dog = plt.imread("dog.jpg")

plt.subplot(1,2,1); plt.imshow(bgr)
plt.subplot(1,2,2); plt.imshow(dog)
print(dog.shape); plt.tight_layout()
```

```python
bgr_size = bgr.shape
same_size_dog = dog[0:bgr_size[0],0:bgr_size[1],:];
plt.imshow(same_size_dog)

blending  = np.linspace(0,1,10);
i=1
for a in blending:
    dst = cv.addWeighted(bgr,a,same_size_dog,1-a,0)
    plt.subplot(2,5,i); plt.axis("off")
    i=i+1
    plt.imshow(dst)
```

```python
import cv2 as cv

bgr_size = bgr.shape
same_size_dog = cv.resize(dog,(bgr_size[1],bgr_size[0]))

plt.subplot(1,3,1); plt.imshow(bgr)
plt.subplot(1,3,2); plt.imshow(same_size_dog)
plt.subplot(1,3,3); plt.imshow(dog)
plt.tight_layout()

blending  = np.linspace(0,1,10);
i=1
for a in blending:
    dst = cv.addWeighted(bgr,a,same_size_dog,1-a,0)
    plt.subplot(2,5,i); plt.axis("off")
    i=i+1
    plt.imshow(dst)
```

## Image Rotation

```python
dog_distorted = cv.resize(dog,(1000,400));
plt.subplot(1,3,1); plt.imshow(dog); plt.axis("off");
plt.subplot(1,3,2); plt.imshow(dog_distorted); plt.axis("off");

# dog_rotated = cv.rotate(dog,cv.ROTATE_180);
dog_rotated = cv.rotate(dog,cv.ROTATE_90_COUNTERCLOCKWISE);
plt.subplot(1,3,3); plt.imshow(dog_rotated); plt.axis("off");
```

## Blurring Images

```python
# Blurring
blur_dog = cv.blur(dog,(100,100))
plt.subplot(1,3,1); plt.imshow(dog); plt.axis("off"); plt.title("Original")
plt.subplot(1,3,2); plt.imshow(blur_dog); plt.axis("off"); plt.title("Blurred")

dog1 = dog.copy();
local_blur = cv.blur(dog[1:1400,1:1000],(100,100));
dog1[1:1400,1:1000] = local_blur
plt.subplot(1,3,3); plt.imshow(dog1); plt.axis("off"); plt.title("Locally Blurred")
```

## Convert Image to Grayscale

```python
# Convert an image to grayscale
gray_dog = cv.cvtColor(dog,cv.COLOR_BGR2GRAY)
# cv.imshow("Gray Dog",gray_dog)
# cv.waitKey(0)

plt.subplot(1,2,1); plt.imshow(dog); plt.axis("off");
plt.subplot(1,2,2); plt.imshow(gray_dog,cmap="Greys"); plt.axis("off");

gray_dog.shape
```

## Image Thresholding

```python
ret, mask = cv.threshold(gray_dog, 165, 255, cv.THRESH_BINARY)

plt.subplot(1,2,1); plt.imshow(gray_dog,cmap="gray",vmin=0,vmax=255); plt.axis("off");
plt.subplot(1,2,2); plt.imshow(mask,cmap="gray",vmin=0,vmax=255); plt.axis("off");

```
