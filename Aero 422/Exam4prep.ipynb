```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# Define the numerator and denominator of the transfer function
# G(s) = 1 / ((s + 1)(s/5 + 1))

# Numerator (constant value in the numerator of the transfer function)
numerator = [1]

# Denominator (coefficients of the denominator of the transfer function)
denominator = [1, 6, 5]  # From the expanded form of (s + 1)(s/5 + 1)

# Create the transfer function
system = signal.TransferFunction(numerator, denominator)

# Frequency range (log scale)
frequencies = np.logspace(-1, 2, 500)  # Frequency from 0.1 to 100 rad/s

# Compute the frequency response
w, mag, phase = signal.bode(system, frequencies)

# Plot Magnitude and Phase
plt.figure(figsize=(10, 6))

# Magnitude plot
plt.subplot(2, 1, 1)
plt.semilogx(w, mag)  # Semilog scale for x-axis (logarithmic)
plt.title('Frequency Response')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Magnitude [dB]')
plt.grid()
# plt.yscale('log')

# Phase plot
plt.subplot(2, 1, 2)
plt.semilogx(w, phase)
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Phase [degrees]')
plt.grid()
# plt.yscale('log')

# Show the plot
plt.tight_layout()
plt.show()
```
